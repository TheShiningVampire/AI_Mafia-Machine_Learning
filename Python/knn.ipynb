```python
x = dfx.values
y = dfy.values.flatten()
```

```python
def KNN(v , k):
    neighbours = []

    for i in range(len(x)):
        point = x[i]
        val = y[i]
        dist = 0
        for j in [0,1]:
            dist += (point[j] - v[j])**2

        dist = np.sqrt(dist)
        neighbours.append((dist , val))


    neighbours = sorted(neighbours)

    count0 = 0

    for i in range(k):
        if (neighbours[i][1] == 0):
            count0 +=1
        
    if (count0>= k/2):
        print('0')
    else:
        print('1')

    print(count0)



```
